### 텍스트 전처리

#### book_intro
- 한글, 영어, 숫자 포함. 그 외 언어는 제외
- 장식 기호 제거
- 반복 구두점 정리
    - ex) “…”, “!!!”
- 연속 공백 정리
- 줄바꿈(”\n”) 제거 후 공백 치환

```python
df[df['book_intro'].apply(lambda x: bool(re.search(r'[^\uAC00-\uD7A3a-zA-Z0-9\s.,!?\'\"()\-]', str(x))))][['isbn', 'title', 'book_intro']].head(5)
```

```python
import re

def clean_text(text):
    if not isinstance(text, str):
        return text
    
    # 줄바꿈 → 공백
    text = re.sub(r'\n+', ' ', text)

    # 한글, 영어, 숫자, 기본 구두점, 공백만 유지
    text = re.sub(r'[^\uAC00-\uD7A3a-zA-Z0-9\s.,!?\'\"()\-]', ' ', text)
    
    # 반복 구두점 정리
    text = re.sub(r'[.]{2,}', '.', text)
    text = re.sub(r'[!]{2,}', '!', text)
    text = re.sub(r'[?]{2,}', '?', text)
    
    # 연속 공백 정리
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text
```

```python
df['book_intro'] = df['book_intro'].apply(clean_text)
df['pub_review'] = df['pub_review'].apply(clean_text)
```

## 알라딘 API를 이용한 카테고리 정보 추가
- API 키 10개 × 5,000회 = 최대 50,000회 호출 가능
- ISBN으로 categoryId, categoryName 조회
- 중간 저장(체크포인트)으로 재시작 시 이어서 진행 가능

```python
import requests
import xml.etree.ElementTree as ET
import os
import time
import pickle
from pathlib import Path
from dotenv import load_dotenv

load_dotenv('../../.env')
API_KEYS = os.getenv('ALADIN_API_KEYS').split(',')

CALLS_PER_KEY = 5000
CHECKPOINT_PATH = '../../data/processed/category_checkpoint.pkl'
OUTPUT_PATH = '../../data/processed/books_with_category.csv'
NS = {'a': 'http://www.aladin.co.kr/ttb/apiguide.aspx'}       # 정상 응답 네임스페이스
NS_ERR = 'http://aladin.co.kr/ttb/apiguide.aspx'              # 에러 응답 네임스페이스 (www. 없음)

os.makedirs('../../data/processed', exist_ok=True)
print(f'처리 대상: {len(df):,}건')
print(f'API 키: {len(API_KEYS)}개 × {CALLS_PER_KEY}회 = 최대 {len(API_KEYS) * CALLS_PER_KEY:,}회')
```

```python
class KeyExhaustedError(Exception):
    pass

def fetch_category(isbn, api_key, retries=3):
    url = 'http://www.aladin.co.kr/ttb/api/ItemLookUp.aspx'
    params = {
        'ttbkey': api_key,
        'itemIdType': 'ISBN',
        'ItemId': str(isbn),
        'output': 'xml',
        'Version': '20131101',
    }
    for attempt in range(retries):
        try:
            resp = requests.get(url, params=params, timeout=15)
            resp.raise_for_status()
            root = ET.fromstring(resp.content)

            # 에러 응답: <error xmlns="http://aladin.co.kr/..."> (www. 없는 네임스페이스)
            if root.tag == f'{{{NS_ERR}}}error':
                code_el = root.find(f'{{{NS_ERR}}}errorCode')
                error_code = code_el.text if code_el is not None else 'unknown'
                if error_code == '8':          # 상품 없음 → 정상 처리
                    return None, None
                else:                          # 그 외(4=잘못된 키 등) → 키 전환
                    raise KeyExhaustedError(f'errorCode={error_code}')

            # 정상 응답
            total = root.find('a:totalResults', NS)
            if total is None or total.text == '0':
                return None, None
            item = root.find('a:item', NS)
            if item is None:
                return None, None
            cat_id = item.find('a:categoryId', NS)
            cat_name = item.find('a:categoryName', NS)
            return (
                cat_id.text if cat_id is not None else None,
                cat_name.text if cat_name is not None else None,
            )
        except KeyExhaustedError:
            raise
        except requests.exceptions.RequestException:
            if attempt < retries - 1:
                time.sleep(2 ** attempt)
        except ET.ParseError:
            return None, None
    return None, None
```

```python
# 체크포인트 로드 (재시작 시 이어서 진행)
if os.path.exists(CHECKPOINT_PATH):
    with open(CHECKPOINT_PATH, 'rb') as f:
        cache = pickle.load(f)
    print(f'체크포인트 로드: {len(cache):,}건 이미 처리됨')
else:
    cache = {}  # {isbn: (categoryId, categoryName)}
    print('체크포인트 없음 — 처음부터 시작')

isbns = df['isbn'].astype(str).tolist()
todo = [isbn for isbn in isbns if isbn not in cache]
print(f'남은 처리: {len(todo):,}건')

call_count = len(cache)
key_idx = min(call_count // CALLS_PER_KEY, len(API_KEYS) - 1)
SAVE_INTERVAL = 500

for i, isbn in enumerate(todo):
    if key_idx >= len(API_KEYS):
        print('모든 API 키 소진 — 중단')
        break

    # 에러 발생 시 즉시 다음 키로 재시도
    while True:
        try:
            cat_id, cat_name = fetch_category(isbn, API_KEYS[key_idx])
            cache[isbn] = (cat_id, cat_name)
            call_count += 1
            break
        except KeyExhaustedError as e:
            print(f'[{call_count:,}회] 키 한도 초과({e}) → 다음 키로 전환')
            key_idx += 1
            if key_idx >= len(API_KEYS):
                print('모든 API 키 소진 — 중단')
                break

    if key_idx >= len(API_KEYS):
        break

    # 카운터 기반 선제 전환 (5000회 도달 전 미리 교체)
    new_key_idx = min(call_count // CALLS_PER_KEY, len(API_KEYS) - 1)
    if new_key_idx != key_idx:
        print(f'[{call_count:,}회] API 키 교체: {API_KEYS[key_idx]} → {API_KEYS[new_key_idx]}')
        key_idx = new_key_idx

    if (i + 1) % SAVE_INTERVAL == 0:
        pct = (i + 1) / len(todo) * 100
        found = sum(1 for v in cache.values() if v[0] is not None)
        print(f'[{i+1:,}/{len(todo):,}] {pct:.1f}% | 조회 성공: {found:,}건 | 사용 키: {key_idx+1}번째')
        with open(CHECKPOINT_PATH, 'wb') as f:
            pickle.dump(cache, f)

with open(CHECKPOINT_PATH, 'wb') as f:
    pickle.dump(cache, f)

found_total = sum(1 for v in cache.values() if v[0] is not None)
print(f'\n완료: 전체 {len(cache):,}건 중 카테고리 조회 성공 {found_total:,}건 ({found_total/len(cache)*100:.1f}%)')
```

```python
# 카테고리 정보를 df에 합치고 CSV 저장
df = df.copy()
df['aladin_category_id'] = df['isbn'].astype(str).map(lambda x: cache.get(x, (None, None))[0])
df['aladin_category_name'] = df['isbn'].astype(str).map(lambda x: cache.get(x, (None, None))[1])

print(f'레코드 수: {len(df):,}')
print(f'카테고리 채워진 행: {df["aladin_category_id"].notna().sum():,}')
print(f'카테고리 빈 행: {df["aladin_category_id"].isna().sum():,}')
print()
print('카테고리 상위 20개:')
print(df['aladin_category_name'].value_counts().head(20))

df.to_csv(OUTPUT_PATH, index=False, encoding='utf-8-sig')
print(f'\n저장 완료: {OUTPUT_PATH}')
```
